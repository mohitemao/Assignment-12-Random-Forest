# Converting  the continuos feature sale into cattegorical feature by grouping as good sale or bad sale

## sales >= 7.496 will be good sale and less than that will be bad sale.

```python
df1['sale']='Good'
df1.head()
```

```python
# split data into dependent and independent data set
x=df1.iloc[:,0:14]
x.head()
```

```python
y=df1.iloc[:,14:]
y
```

```python
# spliting the data into train test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

# building model using random forest

```python
from sklearn.ensemble import RandomForestClassifier as RF
model=RF(n_jobs=4,n_estimators=150,oob_score=True,criterion='entropy')
```

```python
# fit model over train  data set to check accuracy
model.fit(x_train,y_train)
```

```python
model.oob_score_
```

```python
pred_train=model.predict(x_train)
```

```python
# finding accuracy score for train data
from sklearn.metrics import accuracy_score
accuracy_score(y_train,pred_train)
```

```python
#confusion matrix for train data
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,pred_train)
```

```python
# now we will predct data over test dataset
pred_test=model.predict(x_test)
pred_test
```

```python
# finding accuracy score for test data
accuracy_score(y_test,pred_test)
```

```python
#confusion matrix for train data
confusion_matrix(y_test,pred_test)
```
